```python
df.drop(['status_id', 'status_published'], axis=1, inplace=True)
```

```python
df.head(5)
```

```python
X = df
y = df['status_type']
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
X['status_type'] = le.fit_transform(X['status_type'])
y = le.transform(y)
```

```python
X.head()
```

```python
cols = X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

ms = MinMaxScaler()
X = ms.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=[cols])
```

```python
X.head()
```

```python
#with 2 clusters

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
labels = kmeans.labels_

correct_labels = sum(y == labels)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
```

```python
print('Accuracy score: {0:0.2f}%'. format(correct_labels/float(y.size)*100))
```

```python
#Elbow Graph

from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

```python
# with 3 clusters

kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
correct_labels = sum(y == labels)

print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
print('Accuracy score: {0:0.2f}%'. format(correct_labels/float(y.size)*100))
```

```python
# with 4 clusters

kmeans = KMeans(n_clusters=4, random_state=0)

kmeans.fit(X)
labels = kmeans.labels_
correct_labels = sum(y == labels)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
print('Accuracy score: {0:0.2f}%'. format(correct_labels/float(y.size)*100))
```

```python
# with 5 clusters

kmeans = KMeans(n_clusters=5, random_state=0)

kmeans.fit(X)
labels = kmeans.labels_
correct_labels = sum(y == labels)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
print('Accuracy score: {0:0.2f}%'. format(correct_labels/float(y.size)*100))
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X_pca)

plt.figure(figsize=(8, 6))

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

for i in range(len(X_pca)):
    plt.scatter(X_pca[i, 0], X_pca[i, 1], c=colors[kmeans.labels_[i]], s=10)

centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, c='k', label='Centroids')

plt.title('KMeans Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
