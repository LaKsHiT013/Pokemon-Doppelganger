```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.preprocessing import image
from tqdm import tqdm
```

```python
vgg_model = VGG16(weights='imagenet', include_top=False)
```

```python
def load_and_preprocess_image(image_path):
    try:
        img = image.load_img(image_path, target_size=(224, 224))
        img = image.img_to_array(img)
        img = np.expand_dims(img, axis=0)
        img = preprocess_input(img)
        return img
    except Exception as e:
        print(f"Error loading image: {image_path}, Error: {e}")
        return None
```

```python
pokemon_dir = "/kaggle/input/pokemon-images-first-generation17000-files/pokemon"
```

```python
pokemon_species = sorted(os.listdir(pokemon_dir))

species_embeddings = []
species_labels = []
```

```python
for species in tqdm(pokemon_species, desc="Processing Pokemon"):
    species_dir = os.path.join(pokemon_dir, species)
    if not os.path.isdir(species_dir):
        continue
    
    embeddings = []
    
    for img_file in os.listdir(species_dir):
        img_path = os.path.join(species_dir, img_file)
        img = load_and_preprocess_image(img_path)
        if img is None:
            continue
        features = vgg_model.predict(img).flatten()
        embeddings.append(features)
    
    if len(embeddings) == 0:
        print(f"No valid images found for species: {species}")
        continue

    species_embedding = np.mean(embeddings, axis=0)
    species_embeddings.append(species_embedding)
    species_labels.append(species)
```

```python
species_embeddings = np.array(species_embeddings)
species_labels = np.array(species_labels)
```

```python
df = pd.DataFrame(species_embeddings)
df['label'] = species_labels
```

```python
output_file = "pokemon_species_embeddings.csv"
df.to_csv(output_file, index=False)
```
