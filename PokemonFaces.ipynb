```python
import os
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.preprocessing import image
from tqdm import tqdm
```

```python
def create_vgg16_model(input_shape):
    model = Sequential()
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=input_shape))
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2), strides=(2, 2)))

    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2), strides=(2, 2)))

    model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2), strides=(2, 2)))

    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2), strides=(2, 2)))

    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2), strides=(2, 2)))

    model.add(Flatten())
    return model
```

```python
def load_vgg16_weights(model, weights_path):
    model.load_weights(weights_path)
```

```python
def load_and_preprocess_image(image_path):
    try:
        img = image.load_img(image_path, target_size=(224, 224))
        img = image.img_to_array(img)
        img = np.expand_dims(img, axis=0)
        img = preprocess_input(img)
        return img
    except Exception as e:
        print(f"Error loading image: {image_path}, Error: {e}")
        return None
```

```python
weights_path = "/kaggle/input/vggnet/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5"
```

```python
pokemon_dir = "/kaggle/input/pokemon-images-first-generation17000-files/pokemon"
```

```python
pokemon_species = sorted(os.listdir(pokemon_dir))

species_embeddings = []
species_labels = []
```

```python
vgg_model = create_vgg16_model(input_shape=(224, 224, 3))
load_vgg16_weights(vgg_model, weights_path)
```

```python
for species in tqdm(pokemon_species, desc="Processing Pokemon"):
    species_dir = os.path.join(pokemon_dir, species)
    if not os.path.isdir(species_dir):
        continue
    
    embeddings = []
    
    for img_file in os.listdir(species_dir):
        img_path = os.path.join(species_dir, img_file)
        img = load_and_preprocess_image(img_path)
        if img is None:
            continue
        features = vgg_model.predict(img).flatten()
        embeddings.append(features)
    
    if len(embeddings) == 0:
        print(f"No valid images found for species: {species}")
        continue

    species_embedding = np.mean(embeddings, axis=0)
    species_embeddings.append(species_embedding)
    species_labels.append(species)
```

```python
species_embeddings = np.array(species_embeddings)
```

```python
#performing PCA to reduce the dimension to 128 as facial embedings also have a 128 dimensional vector
pca = PCA(n_components=128)
species_embeddings_reduced = pca.fit_transform(species_embeddings)
```

```python
species_labels = np.array(species_labels)
```

```python
df = pd.DataFrame(species_embeddings_reduced)
df['label'] = species_labels
```

```python
output_file = "pokemon_embeddings.csv"
df.to_csv(output_file, index=False)
```
